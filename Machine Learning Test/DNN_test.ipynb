```python
#for data processing
import pandas as pd
import numpy as np
import time
#for MLPClassifier
from sklearn.neural_network import MLPRegressor
#for train_set split
from sklearn.model_selection import train_test_split
#for 정규화 표준화
from sklearn.preprocessing import StandardScaler
#for LGBM
# from lightgbm import LGBMRegressor
#for get mse & R2score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score

#get data from .xlsx
input = pd.read_excel("input_score_last.xlsx")   

#make test_set by .xlsx
y_feature = input['Score']
X_feature = input.drop('Score',axis=1)

#train_set split
X_train, X_test, y_train, y_test = train_test_split(X_feature, y_feature, test_size=0.2, 
random_state =50)

#machine learning using LGBM
# lgb = LGBMRegressor(random_state=0, n_estimators=500, learning_rate=0.05, max_depth= 15, min_child_samples= 10)
# lgb.fit(X_train, y_train)
# y_preds = lgb_reg.predict(X_test)

#machine learning using SVR
print('MLP')
start_time = time.time()

#(6,6) ~ (6,6,6,6,6)
# mlp = MLPRegressor(hidden_layer_sizes=(6,6,6), random_state=42, max_iter=200) #Yes Warning
mlp = MLPRegressor(hidden_layer_sizes=(6,6,6), random_state=42, max_iter=500) #NO Warning

end_time = time.time()
print('Time : {}'.format(end_time-start_time))

#Train data 정규화 및 표준화
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)

# Learn the Regression models
print('Training')
start_time = time.time()
mlp.fit(X_train, y_train)
end_time = time.time()
print('Time : {}'.format(end_time-start_time))

#Test 데이터 표준화 정규화
X_test = scaler.transform(X_test)

# Test model
print('Testing')
start_time = time.time()
y_preds = mlp.predict(X_test)
end_time = time.time()
print('Time : {}'.format(end_time-start_time))

print('Get MSE')
mse = mean_squared_error(y_test,y_preds)
rmse = np.sqrt(mse)

print('\n####After Test####')
print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse,rmse))
print('Variance score(R2score) : {0:.3f}'.format(r2_score(y_test, y_preds)))
```
