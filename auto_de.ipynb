```python
#auto_de_identify.py
import numpy as np
import pandas as pd
import random

make_num = 50

for index in range(0, make_num, 1):
    data = pd.read_csv('sample.csv')
    #데이터 전처리 - 빵구난 부분 처리하기 평균값 넣기 or 지우기?
    data = data.dropna()
    #비식별 방법론 적용
    name = data['이름'].values
    age = data['나이'].values
    phone_num = data['전화번호'].values
    disease = data['병명'].values
    blood = data['혈액형'].values
    smoke = data['흡연여부'].values
    location= data['거주지역'].values
    #이름 비식별화
    #이름 방법론 정하기
    tmp = random.randrange(0,13)
    #전체 데이터에 방법론 적용
    for id in range(0,len(data),1):
        if (tmp == 0):
            pass
        elif (tmp == 1):
            name[id] = name[id].replace(name[id][0], '*')
        elif (tmp == 2):
            name[id] = name[id].replace(name[id][1], '*')
        elif (tmp == 3):
            name[id] = name[id].replace(name[id][2], '*')
        elif (tmp == 4 or tmp == 10 or tmp == 11 or tmp == 12):
            name[id] = name[id].replace(name[id][0:2], '**')
        elif (tmp == 5):
            name[id] = name[id].replace(name[id][1:3], '**')
        elif (tmp == 6):
            name[id] = name[id].replace(name[id][0], '*')
            name[id] = name[id].replace(name[id][2], '*')
        elif (7<=tmp<=9):
            name[id] = name[id].replace(name[id][0:3], '***')

    #나이 비식별화
    #방법론 정하기
    tmp = random.randrange(0,3)
    for id in range(0,len(data),1):
        if (tmp == 0):
            pass
        #일의자리 가림
        elif (tmp == 1):
            data['나이'] = data['나이'].replace([age[id]],[str(age[id])[0:-1]+'*'])
        #다가림
        elif (tmp == 2):
            data['나이'] = data['나이'].replace([age[id]],'**')


    #전화번호 비식별화
    #방법론 고르기
    tmp = random.randrange(0,10)
    if (tmp == 0):
        for id in range(0,len(data),1):
            data['전화번호'] = data['전화번호'].replace([phone_num[id]], ['***-'+phone_num[id][4]+'***-****'])
    elif (tmp == 1):
        for id in range(0,len(data),1):
            data['전화번호'] = data['전화번호'].replace([phone_num[id]], ['***-*'+phone_num[id][5]+'**-****'])
    elif (tmp == 2):
        for id in range(0,len(data),1):
            data['전화번호'] = data['전화번호'].replace([phone_num[id]], ['***-**'+phone_num[id][6]+'*-****'])
    elif (tmp == 3):
        for id in range(0,len(data),1):
            data['전화번호'] = data['전화번호'].replace([phone_num[id]], ['***-***'+phone_num[id][7]+'-****'])
    elif (tmp == 4):
        for id in range(0,len(data),1):
            data['전화번호'] = data['전화번호'].replace([phone_num[id]], ['***-****-'+phone_num[id][9]+'***'])
    elif (tmp == 5):
        for id in range(0,len(data),1):
            data['전화번호'] = data['전화번호'].replace([phone_num[id]], ['***-****-*'+phone_num[id][10]+'**'])
    elif (tmp == 6):
        for id in range(0,len(data),1):
            data['전화번호'] = data['전화번호'].replace([phone_num[id]], ['***-****-**'+phone_num[id][11]+'*'])
    elif (tmp == 7):
        for id in range(0,len(data),1):
            data['전화번호'] = data['전화번호'].replace([phone_num[id]], ['***-****-***'+phone_num[id][12]])
    elif (tmp == 8):
        pass
    else:
        for id in range(0,len(data),1):
            data['전화번호'] = data['전화번호'].replace([phone_num[id]],'***-****-****')

    #나머지 데이터 전체 비식별화
    tmp = random.randrange(0,2)
    for id in range(0,len(data),1):
        if (tmp == 0):
            data['병명'] = data['병명'].replace([disease[id]],['**'])
        else:
            pass

    tmp = random.randrange(0,2)    
    for id in range(0,len(data),1):
        #나머지 데이터 전체 비식별화
        if (tmp == 0):
            data['혈액형'] = data['혈액형'].replace([blood[id]],['**'])
        else:
            pass
        
    tmp = random.randrange(0,2)    
    for id in range(0,len(data),1):
        #나머지 데이터 전체 비식별화
        if (tmp == 0):
            data['거주지역'] = data['거주지역'].replace([location[id]],['**'])
        else:
            pass
    
    tmp = random.randrange(0,2)    
    for id in range(0,len(data),1):
        #나머지 데이터 전체 비식별화
        if (tmp == 0):
            data['흡연여부'] = data['흡연여부'].replace([smoke[id]],['*'])
        else:
            pass  
    print(data)
    #이게 되네       
    data.to_csv("./specimen/specimen_"+str(index)+".csv")
```
